```python
import numpy as np
import matplotlib.pyplot as plt
```

# Espacios muestrales

```python
X2=[2**i for i in range(0,3)]
X6=[6**i for i in range(0,3)]

plt.title('Tamaño de espacio muestral $\Omega $')
plt.plot(X2,label='$2^n$')
plt.plot(X6,label='$6^n$')
plt.legend()
plt.grid(True)
```

```python
Y=[np.exp(x) for x in range(0,10)]
plt.plot(Y)
plt.title('Función exponencial $y(x)=e^x$')
```

# Variables aleatorias

_Ejemplo_: Lanzar una moneda 4 veces. $X$ es la variable aleatoria que cuenta el número de veces que sale sol (1).
$\left\vert \Omega \right\vert=2^4=16$

Vamos a repetir los 4 lanzamientos un total de 20 veces.

```python
X=np.random.randint(0,2,size=4)
```

```python
print('La primera simulación es:',X,'\n','(águila,sol,sol,águila)\n','X=2')
```

```python
Repeticiones=[np.random.randint(0,2,size=4) for i in range(20)]
for i,evento in enumerate(Repeticiones):
    print('ensayo {0}: \t{1}'.format(i+1,evento),', X={}'.format(sum(evento)))
```

```python
Repeticiones=[np.random.randint(0,2,size=4) for i in range(10000)]
X3=[r for r in Repeticiones if sum(r)==3] #Lista con volados donde cayeron 3 soles
```

Si repetimos el experimento $10^4$ veces, en promedio un 25% saldrá 3 veces sol.

```python
len(X3)
```

# Histogramas

```python
N=200000 #número de simulaciones
n=7 #número de volados sucesivos
Simulacion=[np.random.randint(0,2,size=n) for i in range(N)]

#Histograma de la variable aleatoria X=#veces que sale águila
plt.hist([sum(sim) for sim in Simulacion],bins=[i for i in range(n+2)],
         align='left',histtype='barstacked',rwidth=0.95)
```

```python
#Clasificarl los eventos según el número de águilas
X=[[evento for evento in Simulacion if sum(evento)==i] for i in range(n+1)]
```

```python
x=5 #numero de veces que sale aguila (1)
print('De las {0} simulaciones, salieron {1} '
      'veces águilas, lo cual es una fracción de {2:.3f} de todos los eventos'.format(N,len(X[5]),len(X[5])/len(Simulacion)))
```

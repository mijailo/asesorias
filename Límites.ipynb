Definir la sucesión recursiva:
    $$x_{n+1}=\frac{1}{2}\left(x_n+\frac{a}{x_n}\right)$$
donde $a>0$ y $x_0$ es un número real cualquiera, que se elige por conveniencia como cercano a $\sqrt{a}$. La idea es que:
    $$\lim_{n\to\infty} x_n = \sqrt{a}$$

```python
def f(x0,a,n):
    x=x0 
    for i in range(n): 
        x=0.5*(x+a/x) 
    return x
```

```python
X=[f(1,2,i) for i in range(1,20)] 
```

```python
%matplotlib notebook
plt.plot(X,'.',linewidth=5)
plt.grid()
```

Los números del arreglo X, que contiene $x_1,\dots,x_{20}$, se acercan no solamente entre sí, sino también a la raíz de 2.

**Otro ejemplo** La sucesión $a_n=\frac{n}{n+1}$ converge a 1:
    $$\lim_{n\to\infty} a_n = 1$$
lo que significa que, sin importar el "error" $\epsilon>0$, habrá un momento $N\in\mathbb{N}$ en la sucesión a partir del cual los términos $a_{N+1},a_{N+2},\dots$ (es decir, los términos $a_n$ para $n>N$) están a una distancia del 1 menor a $\epsilon$:
    $$\left| a_n - 1 \right|<\epsilon \;\; \forall n>N$$

```python
for i in range(50):
    k=i/(i+1)
    print('a_{0}={1:.3f},\t|a_{0}-1|={2:.3f}'.format(i,k,abs(1-k)))
```

_Conforme n crece, n/(n+1) se acerca a 1 (lo que equivale a decir que la distancia de $a_n$ al 1 tiende a ser más y más chica). Decimos entonces:_
\begin{equation}
\lim_{n\to\infty} \frac{n}{n+1}=1
\end{equation}

1. Primero consideramos la sucesión de funciones: para cada $x\in(0,1)$:
   $$ f_{n}(x):=\sum_{k=0}^n x^k\equiv \frac{1-x^{n+1}}{1-x} $$

**Observación 1** 
    $$\lim_{n\to\infty} x^n=0$$ 

siempre y cuando $x\in(0,1)$.

```python
def pot(x,n):
    return x**n

x=0.9 #En esta x vamos a examinar sus potencias sucesivas
for i in range(1000,1050):
    print('x^{0}={1}'.format(i,pot(x,i)))
```

_Un número positivo menor a 1 ($x\in(0,1)$) elevado a potencias enteras crecientes, tiende a disminuir._

**Observación 2**
        $$\lim_{n\to\infty} f_n(x)=\lim_{n\to\infty} \sum_{k=0}^n x^k=\lim_{n\to\infty} \left( 1+x+x^2+\cdots+x^n\right)=\frac{1}{1-x} \;\; \forall x\in(0,1) $$

_Fijamos x y consideramos $n$ cada vez más grande. La idea es que $f_n(x)$ de x se acerca a $\frac{1}{1-x}$ si $n$ es grande._

```python
def F(x,n): #Aquí defino las funciones f_n(x)
    return sum([x**i for i in range(n+1)])

def lim(x):
    return 1/(1-x)

x=0.9
for i in range(1000,1020): 
    print('f_{} (x)={},\t 1/(1-x)={}'.format(i,F(x,i),lim(x)))
```

**Ver el proceso límite mediante sus gráficas**

```python
%matplotlib notebook
X=np.linspace(0,1,200,endpoint=False)
for i in [1,2,5,10,100,1000]: #graficar f_1(x), f_2(x), ..., f_1000(x)
    plt.plot(X,F(X,i),label='$f_{{{0}}}(x)$'.format(i))
plt.plot(X,lim(X),'.',lw=2,c='blue',label='$1/(1-x)$') #graficar 1/(1-x)
plt.grid()
plt.legend()
plt.show()
```

_Las funciones $f_{n}(x)$ se acercan más y más a $\frac{1}{1-x}$ conforme $n\in\mathbb{N}$ crece. Esto se observa en las líneas continuas, que se acercan más y más al comportamiento de la función representada con puntos azules._
